# 5. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category. What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

# 6. Create a spending flag for each user based on the average price across all their orders.

```python
# Determine mean spending amounts for each customer.

ords_prods_merge['mean_spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check output.

ords_prods_merge.head()
```

```python
# Create flag with 2 types of criteria: <10 = 'Low_spender', >=10 = 'High_spender'

ords_prods_merge.loc[ords_prods_merge['mean_spending'] < 10, 'Spender_type'] = 'Low_spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spending'] >= 10, 'Spender_type'] = 'High_spender'
```

```python
# Check output.

ords_prods_merge['Spender_type'].value_counts(dropna=False)
```

```python
ords_prods_merge.head()
```

# 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. 

```python
# Determine median values for 'days_since_prior_order' column.

ords_prods_merge['median_ord_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check output.

ords_prods_merge.head()
```

```python
# Create flag for reorder frequency with 3 types:
#   >20         = 'Non_frequent customer'
#   >10 & <= 20 = 'Regular customer'
#   <=10        = 'Frequent customer'

ords_prods_merge.loc[ords_prods_merge['median_ord_freq'] > 20, 'customer_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_ord_freq'] > 10) & (ords_prods_merge['median_ord_freq'] <= 20), 'customer_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_ord_freq'] <= 10, 'customer_frequency'] = 'Frequent customer'
```

```python
# Check output.

ords_prods_merge['customer_frequency'].value_counts(dropna=False)
```

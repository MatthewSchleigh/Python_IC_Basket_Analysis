# 3. Create new user-defined if-statement

```python
# Create 'price_label' function for if/else statement.

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return 'Not enough data'
```

```python
# Create 'price_range' column for price_label function output.
# axis=1 refers to rows, axis=0 would mean columns.

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Check the max price value.

df['prices'].max()
```

# 5. If-statements with For-Loops

```python
# Which day are most orders placed?

ords_prods_merge['order_day_of_week'].value_counts(dropna = False)
```

```python
# Create for-loop for defining busiest and least busiest days.

result = []

for value in ords_prods_merge['order_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

# Create a column "Busiest days" for the two busiest and two slowest days.

```python
# On which days are most and least orders placed?

ords_prods_merge['order_day_of_week'].value_counts(dropna = False)
```

```python
# Saturday (0) and Sunday (1) are the two busiest days.
# Tuesday (3) and Wednesday (4) are the two slowest days.
```

```python
# Create for-loop for buesiest and slowest days.

result = []

for value in ords_prods_merge['order_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 1:
        result.append('Busiest day')
    elif value == 3:
        result.append('Slowest day')
    elif value == 4:
        result.append('Slowest day')
    else:
        result.append('Regularly busy')
```

```python
# Create new column 'busiest_days'

ords_prods_merge['busiest_days'] = result
```

# Check the values of this new column for accuracy.

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

# Observations: Orders from the top two busiest days combined are nearer in total to the next three busiest ('regularly busy'), while the slowest two days are just more than half as busy as the three middle days. This suggests that examining both of the top two busiest days would yield better trend data than only the top busiest day, since we can examine a more nuanced relationship to the other five days, for instance the obervation that both of these busy days fall on the weekend.

# Create a new column containing these labels (“Most orders,” “Average orders,” and “Fewest orders”) called “busiest_period_of_day.”

```python
# When are the busiest hours?

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Calculate median and IQR ranges.
```

```python
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False).median()
```

```python
# Using median and IQR ranges, set arguments for Most, Average, and Fewest within for-loop functions.
```

```python
result = []

for value in ords_prods_merge['order_hour_of_day']:
    if value == 10:
        result.append('Most orders')
    elif value == 11:
        result.append('Most orders')
    elif value == 14:
        result.append('Most orders')
    elif value == 15:
        result.append('Most orders')
    elif value == 13:
        result.append('Most orders')
    elif value == 3:
        result.append('Fewest orders')
    elif value == 4:
        result.append('Fewest orders')
    elif value == 2:
        result.append('Fewest orders')
    elif value == 5:
        result.append('Fewest orders')
    elif value == 1:
        result.append('Fewest orders')
    else:
        result.append('Average orders')
```
